### Accessing a specific session

```python
saudi_quali_session = fastf1.get_session(year=2025, gp="jeddah", identifier="Q")
print(f"Session Name: {saudi_quali_session.name}")
print(saudi_quali_session)
```

### Accessing a specific event

```python
# Loading the Saudi 2025 Event all together
saudi_2025 = fastf1.get_event(year=2025, gp="saudiarabia")
print(f"The raceweekend:\n{saudi_2025}")
```

```python
saudi_practice_3 = saudi_2025.get_session(identifier="practice 3")
saudi_quali = saudi_2025.get_session(identifier="qualifying")

print(f"The Practice 3 Session: {saudi_practice_3}")
print(f"The Qualifying Session: {saudi_quali}")
```
